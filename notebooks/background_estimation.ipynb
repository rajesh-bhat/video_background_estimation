```python
import numpy as np
import cv2
from skimage import data, filters
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
# Open Video
cap = cv2.VideoCapture('british_highway_traffic.mp4')

n = 50

# Randomly select n frames
frameIds = cap.get(cv2.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=n)
 
print(frameIds)

# Store selected frames in an array
selected_frames = []
for fid in tqdm(frameIds):
    cap.set(cv2.CAP_PROP_POS_FRAMES, fid)
    ret, frame = cap.read()
    selected_frames.append(frame)

print(np.array(selected_frames).shape)

```

### Background estimation based on median

```python
# Calculate the median along the time axis
median_frame = np.median(selected_frames, axis=0).astype(dtype=np.uint8)
print(median_frame.shape)
plt.imshow(median_frame)
plt.show()
```

### Background estimation based on mean

```python
mean_frame = np.mean(selected_frames, axis=0).astype(dtype=np.uint8)
plt.imshow(mean_frame)
plt.show()
```
